```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
```

```python
# Loading the Dataset
file_path = '/kaggle/input/studentgradesstudy/student-mat.csv'
data = pd.read_csv(file_path, sep=';')
```

```python
# Data Exploration
```

```python
# Checking for missing values
print(data.isnull().sum())
```

```python
# Describing studytime and G3
print(data['studytime'].describe())
print(data['G3'].describe())
```

```python
# Visualizing distributions
plt.figure(figsize=(10, 5))
sns.histplot(data['studytime'], bins=4, kde=False)
plt.title('Distribution of Study Time')
plt.xlabel('Study Time (Categories)')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(10, 5))
sns.histplot(data['G3'], bins=20, kde=True)
plt.title('Distribution of Final Grades (G3)')
plt.xlabel('Final Grades (G3)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Scatter plot for `studytime` vs `G3`
plt.figure(figsize=(10, 5))
sns.scatterplot(x=data['studytime'], y=data['G3'])
plt.title('Study Time vs Final Grades (G3)')
plt.xlabel('Study Time (Categories)')
plt.ylabel('Final Grades (G3)')
plt.show()
```

```python
# Data Cleaning - Outliers
Q1 = data['G3'].quantile(0.25)
Q3 = data['G3'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = data[(data['G3'] < lower_bound) | (data['G3'] > upper_bound)]
cleaned_data = data[(data['G3'] >= lower_bound) & (data['G3'] <= upper_bound)]

print(f"Number of outliers: {len(outliers)}")
```

```python
# Correlation Analysis
correlation = cleaned_data['studytime'].corr(cleaned_data['G3'])
print(f"Correlation between studytime and G3: {correlation}")
```

```python
#  Linear Regression
X = cleaned_data['studytime']
y = cleaned_data['G3']
X = sm.add_constant(X)
model = sm.OLS(y, X).fit()
print(model.summary())
```

```python
#  Visualization of Regression Line
plt.figure(figsize=(10, 5))
sns.scatterplot(x=cleaned_data['studytime'], y=cleaned_data['G3'], label='Data')
plt.plot(cleaned_data['studytime'], model.predict(X), color='red', label='Regression Line')
plt.title('Study Time vs Final Grades (G3) with Regression Line')
plt.xlabel('Study Time (Categories)')
plt.ylabel('Final Grades (G3)')
plt.legend()
plt.show()
```
